```python
import pandas as pd
import numpy as np

path="../../../../../../PROJECT-TEAM-WONDERFUL/subgenre_sample.csv"

csv_df=pd.read_csv(path)

csv_df_split=csv_df['subgenre'].str.split(', ',expand=True)

csv_df_split.columns=['Games','genre1','genre2','genre3']
csv_df_split=csv_df_split.drop('Games',1)

csv_df_split.head()
```

```python
subgenre_list=[]
for i,row in csv_df_split.iterrows():
    for value in row:
        if value not in subgenre_list:
            if value != None:
                subgenre_list.append(value)

subgenre_list.remove('Strategy')
subgenre_list
```

```python
zeros=np.zeros(shape=(len(csv_df_split),len(subgenre_list)))
subgenre_df=pd.DataFrame(zeros,columns=subgenre_list)
subgenre_df.head()
```

```python
for k in range(len(subgenre_list)):
    for i in range(len(csv_df_split)):
        for j in range(len(csv_df_split.columns)):

            if subgenre_list[k] == csv_df_split.iloc[i,j]:
                subgenre_df.iloc[i,k]=1
            else:
                if subgenre_df.iloc[i,k]==1:
                    pass
                else:
                    subgenre_df.iloc[i,k]=0
subgenre_dummies=subgenre_df.iloc[:,:-2]

subgenre_dummies
```

```python
subgenre_dummies.to_csv('subgenre_dummies.csv',index=False)
```
